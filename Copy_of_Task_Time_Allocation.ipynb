```python
import pandas as pd
import numpy as np
import random

# Mock data for projects and tasks
projects = ['A', 'B', 'C', 'D', 'E']
num_tasks = np.random.randint(5, 9, size=len(projects))
total_time = np.random.randint(12, 17, size=len(projects))

data = {
    'ProjectID': (projects),
    'TotalTime': (total_time),
}

# Create a DataFrame
df = pd.DataFrame(data)
df
```

```python
# Add 5 to 8 tasks randomly for each project
df['NumTasks'] = np.random.randint(5, 9, size=len(df))

# Initialize an empty DataFrame for tasks
df_tasks = pd.DataFrame()
df
```

```python
# Allocate time for each task within each project
for _, row in df.iterrows():
    project_id = row['ProjectID']
    total_time_project = row['TotalTime']
    num_tasks = row['NumTasks']

    # Randomly allocate time for each task
    task_times = np.random.uniform(0.1, 5, size=num_tasks)

    # Normalize task times to ensure the sum equals the total time of the project
    task_times = (task_times / np.sum(task_times)) * total_time_project

    # Round the times to two decimal places
    task_times = np.round(task_times, 2)

    # Create a DataFrame for tasks in the current project
    df_tasks = pd.concat([df_tasks, pd.DataFrame({'ProjectID': [project_id] * num_tasks, 'TaskID': range(1, num_tasks + 1), 'AllocatedTime': task_times})])

```

```python
# Merge the project and task DataFrames
df = pd.merge(df_tasks, df[['ProjectID', 'TotalTime']], on='ProjectID')
df
```

```python
# Calculate the value percentage for each task and project
df['ValuePercentage_Task'] = ((df['AllocatedTime'] / df['TotalTime']) * 100).round(1)
df['ValuePercentage_Project'] = df.groupby('ProjectID')['ValuePercentage_Task'].transform('sum').round(1)
df
```

```python
# Display the final DataFrame
df_dict = df.to_dict(orient='records')
df_dict
```

```python
# Calculate the average task time per project
average_time_per_project = df.groupby('ProjectID')['AllocatedTime'].mean().round(1)

# Convert the result to a dictionary
average_time_per_project_dict = average_time_per_project.to_dict()

# Display the average task time per project
print("Average Task Time per Project:")
print(average_time_per_project_dict)
```
