```python
# Generate data for task_df
num_tasks = 10

task_data = {
    'task_id': list(range(1, num_tasks + 1)),
    'start_date': [datetime(2024, 1, 1) + timedelta(days=random.randint(1, 30)) for _ in range(num_tasks)],
}

# Ensure that the difference in days is at least 1 and does not exceed 5
task_data['end_date'] = [min(start_date + timedelta(days=random.randint(1, 5)), datetime(2024, 1, 31)) for start_date in task_data['start_date']]

# Calculate hours_spent (not exceeding 40 and not zero)
task_data['hours_spent'] = [
    max(min((end_date - start_date).days * 8, 40), 8)  # Ensuring at least 1 day of work (8 hours)
    for start_date, end_date in zip(task_data['start_date'], task_data['end_date'])
]

# Generate task_revenue
task_data['task_revenue'] = [random.randint(500, 1000) for _ in range(num_tasks)]

# Create task_df DataFrame
task_df = pd.DataFrame(task_data)

task_df.head(10)
```

```python
# Generate task assignments
task_assignments = []
for emp_id in emp_df['employee_id']:
    num_assigned_tasks = random.randint(1, min(5, num_tasks))  # Assign at most 5 tasks
    assigned_tasks = random.sample(range(1, num_tasks + 1), num_assigned_tasks)
    task_assignments.append((emp_id, assigned_tasks))

# Create a DataFrame for task assignments
task_assignments_df = pd.DataFrame(task_assignments, columns=['employee_id', 'tasks'])

# Merge task_assignments_df with emp_df
emp_df = pd.merge(emp_df, task_assignments_df, on='employee_id', how='left')

# Fill NaN values in 'tasks' column with an empty list
emp_df['tasks'] = emp_df['tasks'].apply(lambda x: x if isinstance(x, list) else [])

# Add 'tasks_completed' column
emp_df['tasks_completed'] = emp_df['tasks'].apply(len)

emp_df.head(10)
```

```python
# Merge emp_df with task_df to get 'hours_spent' for each task
merged_df = emp_df.explode('tasks').merge(task_df, left_on='tasks', right_on='task_id', how='left')

# Calculate salary for each row
merged_df['salary'] = merged_df['emp_wages'] * merged_df['hours_spent']

# Group by 'employee_id' and sum the salaries for each employee
salary_df = merged_df.groupby('employee_id')['salary'].sum().reset_index()

# Merge the calculated salaries back to emp_df
emp_df = pd.merge(emp_df, salary_df, on='employee_id', how='left')


emp_df.head(10)
```

```python
#Profit per Employee

# Calculate profit for each employee
emp_df['profit'] = 0  # Initialize the 'profit' column with zeros

for index, row in emp_df.iterrows():
    task_ids = row['tasks']

    # Calculate profit for the employee
    task_revenues = task_df[task_df['task_id'].isin(task_ids)]['task_revenue']
    emp_profit = sum(row['emp_wages'] * task_revenues) - row['salary']

    # Update the 'profit' column in emp_df
    emp_df.at[index, 'profit'] = emp_profit

emp_df.head(10)
```

```python

# Initialize an empty dictionary to store task_executor lists
task_executor_dict = {}

# Iterate through emp_df to populate task_executor_dict
for index, row in emp_df.iterrows():
    for task_id in row['tasks']:
        if task_id not in task_executor_dict:
            task_executor_dict[task_id] = []
        task_executor_dict[task_id].append(row['employee_id'])

# Create a new column 'task_executor' in task_df
task_df['task_executor'] = task_df['task_id'].map(task_executor_dict)

task_df.head(10)
```

```python
# Calculate profit for each task
task_df['profit'] = 0  # Initialize the 'profit' column with zeros

for index, row in task_df.iterrows():
    task_executor_list = row['task_executor']

    # Ensure 'task_executor_list' is a list
    if not isinstance(task_executor_list, list):
        task_executor_list = [task_executor_list]

    # Calculate profit for the task
    task_profit = row['task_revenue'] - sum(emp_df[emp_df['employee_id'].isin(task_executor_list)]['salary'])

    # Update the 'profit' column in task_df
    task_df.at[index, 'profit'] = task_profit

task_df.head(10)
```
