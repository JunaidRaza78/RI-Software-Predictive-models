```python
import pandas as pd
import numpy as np
from faker import Faker
import random

# Create a Faker instance
fake = Faker()

# Set seed for reproducibility
random.seed(42)

# Function to generate random barcode
def generate_barcode():
    return ''.join(random.choice('0123456789') for i in range(12))

# Function to generate mock product data
def generate_mock_product_data(num_records, num_months):
    data = {
        'id': list(range(1, num_records + 1)),
        'minimal_amount': [random.randint(1, 10) for _ in range(num_records)],
        'price': [round(random.uniform(10, 1000), 2) for _ in range(num_records)],
        'name': [fake.word() for _ in range(num_records)],
        'avatar': [None] * num_records,
        'company_id': [random.choice([1, 2, 3]) for _ in range(num_records)],
        'relationship_id': [random.choice([1, 2, 3]) for _ in range(num_records)],
        'description': [fake.text() for _ in range(num_records)],
        'barcode': [generate_barcode() for _ in range(num_records)],
        'currency': [random.choice(['USD', 'EUR', 'UAH']) for _ in range(num_records)],
        'sales': [random.randint(50, 500) for _ in range(num_records)],
        'rating': [random.uniform(1, 5) for _ in range(num_records)],
        'Month': np.tile(np.arange(1, num_months + 1), int(np.ceil(num_records / num_months)))[:num_records],
    }
    return pd.DataFrame(data)

# Function to generate mock service data
def generate_mock_service_data(num_records, num_months):
    data = {
        'id': list(range(1, num_records + 1)),
        'company_id': [random.choice([1, 2, 3]) for _ in range(num_records)],
        'relationship_id': [random.choice([1, 2, 3]) for _ in range(num_records)],
        'avatar': [None] * num_records,
        'name': [fake.word() for _ in range(num_records)],
        'price': [round(random.uniform(10, 1000), 2) for _ in range(num_records)],
        'description': [fake.text() for _ in range(num_records)],
        'barcode': [generate_barcode() for _ in range(num_records)],
        'currency': [random.choice(['USD', 'EUR', 'UAH']) for _ in range(num_records)],
        'sales': [random.randint(50, 500) for _ in range(num_records)],
        'rating': [random.uniform(1, 5) for _ in range(num_records)],
        'Month': np.tile(np.arange(1, num_months + 1), int(np.ceil(num_records / num_months)))[:num_records],
    }
    return pd.DataFrame(data)

# Generate mock product and service data
num_product_records = 100
num_service_records = 100
num_months = 12

products_df = generate_mock_product_data(num_product_records, num_months)
services_df = generate_mock_service_data(num_service_records, num_months)

products_df.head()


```

```python
def analyze_product_performance(product_data):
    """
    Analyze the performance of a product.

    Parameters:
    - product_data: DataFrame containing data for a single product.

    Returns:
    - None (prints analysis results).
    """

    # Calculate key performance indicators
    total_sales = product_data['sales'].sum()
    average_price = product_data['price'].mean()
    total_revenue = total_sales * average_price
    satisfaction_score = product_data['rating'].mean()

    # Print analysis results
    print(f"Product Analysis for {product_data['name'].iloc[0]} (Product ID: {product_data['id'].iloc[0]})")
    print("------------------------------------------")
    print(f"Total Sales: {total_sales}")
    print(f"Average Price: {average_price:.2f}")
    print(f"Total Revenue: {total_revenue:.2f}")
    print(f"Average Satisfaction Score: {satisfaction_score:.2f}")

analyze_product_performance(products_df)
```

```python
def find_best_performing_product(all_products_data):
    """
    Find the best-performing product based on total revenue.

    Parameters:
    - all_products_data: DataFrame containing data for all products.

    Returns:
    - DataFrame with information about the best-performing product.
    """

    # Calculate total revenue for each product
    all_products_data['TotalRevenue'] = all_products_data['sales'] * all_products_data['price']

    # Find the product with the highest total revenue
    best_product = all_products_data.loc[all_products_data['TotalRevenue'].idxmax()]

    return best_product


best_performing_product = find_best_performing_product(products_df)

# Print information about the best-performing product
print("Best Performing Product:")
print(best_performing_product)
#df_dict = best_performing_product.to_dict(orient='records')

# Display the dictionary
#print(df_dict)
```

```python
def find_worst_performing_product(all_products_data):
    """
    Find the worst-performing product based on total revenue.

    Parameters:
    - all_products_data: DataFrame containing data for all products.

    Returns:
    - DataFrame with information about the worst-performing product.
    """

    # Calculate total revenue for each product
    all_products_data['TotalRevenue'] = all_products_data['sales'] * all_products_data['price']

    # Find the product with the lowest total revenue
    worst_product = all_products_data.loc[all_products_data['TotalRevenue'].idxmin()]

    return worst_product

worst_performing_product = find_worst_performing_product(products_df)

# Print information about the worst-performing product
print("Worst Performing Product:")
worst_performing_product
```

```python
def analyze_performance_factors(product_data, performance_type='high'):
    """
    Analyze factors contributing to the performance of a product.

    Parameters:
    - product_data: DataFrame containing data for a single product.
    - performance_type: 'high' or 'low' to specify whether to analyze high or low-performing product.

    Returns:
    - DataFrame with information about the factors contributing to performance.
    """

    # Sort product data based on total revenue
    sorted_product_data = product_data.sort_values(by='TotalRevenue', ascending=(performance_type == 'low'))

    # Select the top-performing or bottom-performing product
    selected_product = sorted_product_data.iloc[0]

    # Extract factors contributing to performance
    factors_contributing = {
        'ProductID': selected_product['id'],
        'Name': selected_product['name'],
        'TotalRevenue': selected_product['TotalRevenue'],
        'Sales': selected_product['sales'],
        'Price': selected_product['price'],
        'rating': selected_product['rating']
    }

    return factors_contributing

high_performance_factors = analyze_performance_factors(products_df, performance_type='high')
low_performance_factors = analyze_performance_factors(products_df, performance_type='low')

# Print information about factors contributing to high-performing product
print("Factors Contributing to High-Performing Product:")
print(high_performance_factors)

# Print information about factors contributing to low-performing product
print("\nFactors Contributing to Low-Performing Product:")
print(low_performance_factors)
```

```python
import pandas as pd

def product_performance_dict(df):
    """
    Generate a nested dictionary with product performance data for each product in the desired format.

    Parameters:
    - df: DataFrame containing mock product data with columns 'id', 'minimal_amount', 'price', 'sales', 'rating', 'Month'

    Returns:
    - product_performance_dict: Nested dictionary with product performance data for each product
    """
    # Convert 'Month' to timestamps
    df['Month'] = pd.to_datetime(df['Month'], format='%m', errors='coerce')

    # Create a nested dictionary for product performance
    product_performance_dict = {}

    for product_id, data in df.groupby('id'):
        key = f'Product_ID_{product_id}_rating'
        product_performance_dict[key] = []

        for idx, row in data.iterrows():
            product_performance_dict[key].append({
                'Time': row['Month'],
                'Rating': row['rating'],
            })

    return product_performance_dict


product_performance_dict = product_performance_dict(products_df)
print(product_performance_dict)
```
