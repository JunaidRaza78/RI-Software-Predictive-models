```python
import pandas as pd

def generate_code_single(purpose, age_start, age_end, gender):
    purpose = purpose[0] + purpose.split()[1][0]  # Extract the first letter of the purpose and the first letter after the space
    code = f"{purpose}{age_start}to{age_end}{gender[0]}"
    return code

# Read data from Excel file
data = pd.read_excel("/content/new model data.xlsx")

# Generate codes for each row
generated_codes = []
for index, row in data.iterrows():
    code = generate_code_single(row['purpose'], row['age_start'], row['age_end'], row['gender'])
    generated_codes.append(code)

# Write generated codes to a new Excel file
generated_data = pd.DataFrame({"Generated Code": generated_codes})
generated_data.to_excel("generated_codes.xlsx", index=False)

print("Generated codes saved to 'generated_codes.xlsx'")
```
